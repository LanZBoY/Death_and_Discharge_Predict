```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date, timedelta
from tqdm import tqdm
from random import randint
```

```python
dataframe_time = pd.read_csv('./data_20221229_timeseries.csv')
```

```python
dataframe_time.columns
```

```python
select_id = dataframe_time['stay_id'][randint(0, len(dataframe_time['stay_id'].unique()))]
stay_sample = dataframe_time[dataframe_time['stay_id'] == select_id]
stay_sample.sort_values(axis=0, by = 'to_date')
```

```python
len(dataframe_time['stay_id'].unique())
```

```python
cotineous = True
not_cotineous_list = []
for stay_id in tqdm(dataframe_time['stay_id'].unique()):
    stay = dataframe_time[dataframe_time['stay_id'] == stay_id]['to_date']
    stay = stay.sort_values(axis = 0)
    stay = stay.reset_index().drop(columns = ['index'])
    for i in range(1, len(stay)):
        delta = pd.to_datetime(stay['to_date'][i]) - pd.to_datetime(stay['to_date'][i - 1])
        if delta != timedelta(days=1):
            not_cotineous_list.append(stay_id)
            cotineous = False
            break
```

```python
len(not_cotineous_list)
```
